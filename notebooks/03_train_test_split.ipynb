# Phase 4: Train/Test Split & Baseline

In this notebook, we perform a time-aware split of our data and establish a baseline performance metric. This ensures our model evaluation is realistic and not biased by 'shuffling' time-series data.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score
import os

df = pd.read_csv('../dataset/processed/cleaned_traffic_data.csv')
print(f"Total rows: {len(df)}")
```

## 1. Chronological Split
We use 70% for training, 15% for validation, and 15% for testing. 

```python
# Calculate split points
train_size = int(len(df) * 0.7)
val_size = int(len(df) * 0.15)

train_df = df.iloc[:train_size]
val_df = df.iloc[train_size : train_size + val_size]
test_df = df.iloc[train_size + val_size :]

print(f"Train size: {len(train_df)}")
print(f"Validation size: {len(val_df)}")
print(f"Test size: {len(test_df)}")
```

## 2. Establish Baseline
Our baseline will be the global average traffic volume from the training set. 

```python
baseline_prediction = train_df['traffic_volume'].mean()
print(f"Baseline (Mean Traffic): {baseline_prediction:.2f} vehicles/hour")

# Evaluate baseline on validation set
y_val = val_df['traffic_volume']
baseline_preds = [baseline_prediction] * len(y_val)

mae = mean_absolute_error(y_val, baseline_preds)
print(f"Baseline MAE on Validation Set: {mae:.2f}")
```

## 3. Save Splits for Training
We save these as separate files to make Phase 5 (Model Training) cleaner.

```python
os.makedirs('../dataset/processed/splits', exist_ok=True)

train_df.to_csv('../dataset/processed/splits/train.csv', index=False)
val_df.to_csv('../dataset/processed/splits/val.csv', index=False)
test_df.to_csv('../dataset/processed/splits/test.csv', index=False)

print("Success! Data splits saved to dataset/processed/splits/")
```
